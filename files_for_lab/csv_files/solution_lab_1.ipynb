```python
def fix_value(value):
    if isinstance(value, str):
        if '%' in value:
            return float(value.replace('%',''))*100
        elif value.lower() == 'nan':
            return None
        else:
            return float(value)
    else:
        return value


new_data['customer lifetime value'] = new_data['customer lifetime value'].apply(fix_value)
```
